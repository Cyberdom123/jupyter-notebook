```python
from sympy import *
import numpy as np
from sympy import symbols, cos, sin
from sympy.plotting import plot_parametric
u = symbols("u")
y = cos(u)
g = y.diff(u)

p = plot_parametric((u, y, (u, -7, 7)), (u, g, (u, -7, 7)))

p[0].label = "cos(x)"
p[1].label = "d(cos(x)) = -sin(x)"
p.legend = True
p[1].line_color = 'red'
p.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(1, 5, 25)
y = x**2
z = x**3
g = y + z

plt.stem(x, y, label = 'x^2')
plt.stem(x, z, label = 'x^3')
plt.plot(x, g, color="red", label='x^2 + x^3')

plt.legend()
plt.show()
```

```python
from sympy import *
from sympy.plotting import plot_parametric

x = sympy.Symbol('x')
y = exp(3*x)
yprime = y.diff(x)
print(f'f: {y}')
print (f'Derivative of f: {yprime}')

p = 2 #tangent point
y1 = y.subs({x:p})
a = yprime.subs({x:p}) 
b = y1 - a*p

g = a*x + b #tangent lines to function

print(f"\n a: {a},\n b : {b},\n g : {g}")


p = plot_parametric((x, y, (x, -2, 2.5)), (x, g, (x, 1, 2.5))) #fix plz
p[1].line_color = 'red'
p.show()
```
