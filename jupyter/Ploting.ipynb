```python
import matplotlib.pyplot as plt
import numpy as np

def T(n):
    if n == 1:
        return 100
    else:
        return np.random.choice([-1,1]) + T(n-1)
print([T(i) for i in range(1,50)])
fig, ax = plt.subplots()  # Create a figure containing a single axes.

ax.plot([1,2,3,4], [T(i) for i in range(1,5)])
```

```python
x = np.linspace(0, 2, 100)

# Note that even in the OO-style, we use `.pyplot.figure` to create the figure.
fig, ax = plt.subplots()  # Create a figure and an axes.
ax.plot(x, x, label='linear')  # Plot some data on the axes.
ax.plot(x, x**2, label='quadratic')  # Plot more data on the axes...
ax.plot(x, x**3, label='cubic')  # ... and some more.
ax.set_xlabel('x label')  # Add an x-label to the axes.
ax.set_ylabel('y label')  # Add a y-label to the axes.
ax.set_title("Simple Plot")  # Add a title to the axes.
ax.legend()  # Add a legend.
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
import sys
sys.setrecursionlimit(10000)
plt.rcParams['figure.figsize'] = [15, 10]

# Fixing random state for reproducibility
np.random.seed(19680801)

fig, ax = plt.subplots()

def T(n):
    if n == 1:
        return 100
    else:
        return (np.random.choice([-1,1]) + T(n-1))
print([T(i) for i in range(1,50)])
    
x = np.arange(0.0, 5000.0, 1.0)
y = [T(i) for i in range(1,5001)]
s = 50
plt.scatter(x, y, s, c="b", marker=r'.',
            label="movement for the number of steps")

def constant_function(x):
    return np.full(x.shape, 100)
t1 = np.arange(0.0,5000.0,1.0)
plt.plot(t1, constant_function(t1))

plt.ylabel("Position")
plt.xlabel("Step")
plt.legend(loc='upper left')

plt.show()
```
